```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm, uniform 
import seaborn as sns
```

```python
# question 1
n = 100
p = 0.09 
k = np.arange(0, 101)
binomial = binom.pmf(k=k, n=n, p=p)
bar = plt.bar(x=k, height=binomial)
plt.xlim(0,50)
plt.xlabel('number of doses')
plt.ylabel('probability')
for i in range(0,4):
    bar[i].set_color('r')
plt.show()
three = binomial[3]
at_most_three = sum(binomial[:4])
print(f'probability that only 3 will fail: {three}')
print(f'at most three will fail: {at_most_three}')
```

```python
# question 2
n = 200
p = 0.09 
k = np.arange(0, 201)
binomial = binom.pmf(k=k, n=n, p=p)
bar = plt.bar(x=k, height=binomial)
plt.xlim(0,50)
plt.xlabel('number of doses')
plt.ylabel('probability')
for i in range(30, 200):
    bar[i].set_color('r')
plt.show()
at_least_30= 1 - sum(binomial[:30])
print(f'probability that 30 or more will fail: {at_least_30}')
```

```python
doses = pd.read_csv('doses.csv')
mean = doses['time_of_effect'].mean()
sigma = doses['time_of_effect'].std()
sns.displot(doses['time_of_effect'], kde=True)

eleven_point_5 = norm.cdf(x=11.5, loc=mean, scale=sigma)
top_10 = norm.ppf(q=.9, loc=mean, scale=sigma)
se = sigma / np.sqrt(50)
interval = norm.interval(confidence=.95, loc=mean, scale=se)
print(f'probability that it will be less than 11.5: {eleven_point_5}')
print(f'90th percentile: {top_10}')
print(f'95% confidence range: {interval}')
```
